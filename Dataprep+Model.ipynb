**Pipeline**

```python
all_numeric_columns = list(df.select_dtypes(['int64', 'float64']).columns)
all_categorical_columns = list(df.select_dtypes('object').columns)
```

```python
all_numeric_columns
```

```python
all_categorical_columns
```

```python
all_numeric_columns.remove('Production cost')
```

```python
set_config(display="diagram")
```

```python
numeric_transformer = Pipeline(steps=[
    ('impmean', SimpleImputer(strategy='mean')),
    ('outlier_remove', Winsorizer(capping_method='quantiles', tail='both', fold=0.05)),
    ('num', StandardScaler()),
])

categorical_transformer = Pipeline(steps=[
    ('impmode', SimpleImputer(strategy='most_frequent')),
    ('cat', OneHotEncoder()),
])

# Define the column transformer with the preprocessing steps for each column type
preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, all_numeric_columns),
    ('cat', categorical_transformer, all_categorical_columns)
], sparse_threshold=0, remainder='passthrough')

preprocessor
```

```python
model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', XGBRegressor())
])
model_pipeline
```

```python
y_pipeline = df['Production cost']
y_pipeline
```

```python
X_pipeline = df.drop(['Production cost'], axis=1)
X_pipeline.head()
```

```python
X_pipeline_train, X_pipeline_test, y_pipeline_train, y_pipeline_test = train_test_split(X_pipeline, y_pipeline, train_size=0.75, test_size=0.25)
```

```python
X_pipeline_train.columns
```

```python
model_pipeline.fit(X_pipeline_train, y_pipeline_train)
```

```python
y_pipeline_train_pred = model_pipeline.predict(X_pipeline_train)
rmse_pipeline_train_pred = np.sqrt(mean_squared_error(y_pipeline_train, y_pipeline_train_pred))
rmse_pipeline_train_pred
```

```python
y_pipeline_test_pred = model_pipeline.predict(X_pipeline_test)
rmse_pipeline_test_pred = np.sqrt(mean_squared_error(y_pipeline_test, y_pipeline_test_pred))
rmse_pipeline_test_pred
```
